Your task is to do the following with this data:
1. Load the ratings.dat file
2. Merge information from the movies.dat file to include movie titles and genres
    * Separate the genres using get_dummies function
3. Merge information from the users.dat file to include gender, age, and occupation for the users giving the ratings

Use this data to find:
1. Average rating for each genre
2. Average age for each genre
3. Count of genders for each genre
4. Count of user occupations for each genre

```python
import pandas as pd 

ratings = pd.read_csv("https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/movielens/ratings.dat", delimiter="::", names=['UserId', 'MovieId', 'Rating', 'Timestamp'])
ratings
```

```python
movies = pd.read_csv("https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/movielens/movies.dat", delimiter="::", names=['MovieId', 'Title', 'Genres'])
movies
```

```python
movie_genres = movies['Genres'].str.get_dummies("|")
movie_genres
```

```python
movies = movies.merge(movie_genres, left_index=True, right_index=True, how='outer')
movies
```

```python
ratings = pd.merge(ratings, movies, how='outer', left_on='MovieId', right_on='MovieId')
```

```python
users = pd.read_csv('https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/datasets/movielens/users.dat', delimiter='::', names=['UserId', 'Gender', 'Age', 'Occupation', 'Zip-Code'])
users
```

```python
ratings = ratings.merge(users, left_on='UserId', right_on='UserId', how='outer')
ratings
```

Use this data to find:
1. Average rating for each genre
2. Average age for each genre
3. Count of genders for each genre
4. Count of user occupations for each genre

```python
categories = ['Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']

for i in categories:    
    print(f"{i}", ratings[ratings[i] == 1]['Rating'].mean())

```
