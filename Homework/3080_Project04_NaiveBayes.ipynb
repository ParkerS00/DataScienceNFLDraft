The Naive Bayes algorithm utilized Bayes' Theorem:
$$P(Hypothesis | Evidence) = \frac{P(Hypothesis) P(Evidence | Hypothesis)}{P(Evidence)}$$

To train this model, we take the relative frequency for each Hypothesis and the corresponding evidence and use those as the probabilities.

Let's train this model. You have the following dataset:

```python
import numpy as np
import pandas as pd
```

```python
fruit = pd.read_csv('https://raw.githubusercontent.com/drolsonmi/math3080/main/Datasets/fruit.csv')
fruit
```

```python
print(fruit['Fruit'].unique())
print(fruit['Color'].unique())
print(fruit['Size'].unique())
# print(fruit[fruit['Fruit'] == 'Lime'])
# print(fruit[fruit['Fruit'] == 'Apple'])
# print(fruit[fruit['Fruit'] == 'Peach'])
```

```python
# P(Apple | Red & Small) = ( P(Apple) * P(Red & Small | Apple) ) / P (Red & Small)
PA = (fruit['Fruit'] == 'Apple').sum() / len(fruit)
PRSA = ((fruit['Size'] == 'Small') & (fruit['Color'] == 'Red') & (fruit['Fruit'] == 'Apple')).sum() / (fruit['Fruit'] == 'Apple').sum()
PRS = ((fruit['Size'] == 'Small') & (fruit['Color'] == 'Red')).sum() / len(fruit)

Prob = (PA * PRSA) / PRS
Prob
```

```python
# P(Apple | Green) = P(Apple) * P(Green | Apple) / P(Green)
PA = (fruit['Fruit'] == 'Apple').sum() / len(fruit)
PGA = ((fruit['Color'] == 'Green') & (fruit['Fruit'] == 'Apple')).sum() /  (fruit['Fruit'] == 'Apple').sum()
PG = (fruit['Color'] == 'Green').sum() / len(fruit)

PAG = (PA * PGA) / PG
print(PAG)

# P(Lime | Green) = P(Lime) * P(Green | Lime) / P(Green)
PA = (fruit['Fruit'] == 'Lime').sum() / len(fruit)
PGL = ((fruit['Color'] == 'Green') & (fruit['Fruit'] == 'Lime')).sum() /  (fruit['Fruit'] == 'Lime').sum()
PG = (fruit['Color'] == 'Green').sum() / len(fruit)

PLG = (PA * PGL) / PG
print(PLG)

# P(Peach | Green) = P(Peach) * P(Green | Peach) / P(Green)
PA = (fruit['Fruit'] == 'Peach').sum() / len(fruit)
PGP = ((fruit['Color'] == 'Green') & (fruit['Fruit'] == 'Peach')).sum() /  (fruit['Fruit'] == 'Peach').sum()
PG = (fruit['Color'] == 'Green').sum() / len(fruit)

PPG = (PA * PGP) / PG
print(PPG)
```

```python
titanic.isna().sum()
titanic.drop('deck', axis=1, inplace=True)
titanic.drop('age', axis=1, inplace=True)
titanic.head()
```

```python
# P(Survived | Sex, Class, Alone) = P(Survived) * P(Sex & Class & Alone | Survived) / P (Sex & Class & Alone)
# should be 
def probability(Survived, Sex, Class, Alone):
    PS = (titanic['survived'] == Survived).sum() / len(titanic['survived'])
    PSCAS = ((titanic['sex'] == Sex) & (titanic['class'] == Class) & (titanic['alone'] == Alone) & (titanic['survived'] == Survived)).sum() / (titanic['survived'] == Survived).sum()
    PSCA = ((titanic['sex'] == Sex) & (titanic['class'] == Class) & (titanic['alone'] == Alone)).sum() / len(titanic['survived'])

    return (PS * PSCAS) / PSCA

print(probability(1, 'male', 'Third', False))
print(probability(0, 'female', 'Third', True))
```
