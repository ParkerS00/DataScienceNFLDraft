# Top 5 Correlated
1. radius_mean to perimeter_mean
2. perimeter_worst to radius_worst
3. area_mean to radius_mean
4. area_mean to perimeter_mean
5. radius_worst to area_worst

```python
fig, ax = plt.subplots(5, 1, figsize=(10, 50))

sns.scatterplot(data=data, x=data['radius_mean'], y=data['perimeter_mean'], ax=ax[0])
ax[0].set_title('Average radius vs the average perimeter')

sns.scatterplot(data=data, x=data['perimeter_worst'], y=data['radius_worst'], ax=ax[1])
ax[1].set_title('Worst perimiter vs the worst radius')

sns.scatterplot(data=data, x=data['area_mean'], y=data['radius_mean'], ax=ax[2])
ax[2].set_title('Average area vs the average radius')

sns.scatterplot(data=data, x=data['area_mean'], y=data['perimeter_mean'], ax=ax[3])
ax[3].set_title('Average area vs the average perimeter')

sns.scatterplot(data=data, x=data['radius_worst'], y=data['area_worst'], ax=ax[4])
ax[4].set_title('Worst radius vs the worst area')
```

```python
def calc_b1(X, Y):
    xbar = X - X.mean()
    ybar = Y - Y.mean()

    return (xbar.dot(ybar))/(xbar.dot(xbar))

def calc_b0(X, Y):
    b1 = calc_b1(X, Y)

    return (Y.mean() - b1 * X.mean())
```

```python
# radius_mean to perimeter_mean
b1_1 = calc_b1(data['radius_mean'], data['perimeter_mean'])
b0_1 = calc_b0(data['radius_mean'], data['perimeter_mean'])

print(f"y = {b0_1:0.3f} + {b1_1:0.3f}x")

# perimeter_worst to radius_worst
b1_2 = calc_b1(data['perimeter_worst'], data['radius_worst'])
b0_2 = calc_b0(data["perimeter_worst"], data['radius_worst'])

print(f"y = {b0_2:0.3f} + {b1_2:0.3f}x")

# area_mean to radius_mean
b1_3 = calc_b1(data['area_mean'], data['radius_mean'])
b0_3 = calc_b0(data['area_mean'], data['radius_mean'])

print(f"y = {b0_3:0.3f} + {b1_3:0.3f}x")

# area_mean to perimeter_mean
b1_4 = calc_b1(data['area_mean'], data['perimeter_mean'])
b0_4 = calc_b0(data['area_mean'], data['perimeter_mean'])

print(f"y = {b0_4:0.3f} + {b1_4:0.3f}x")

# radius_worst to area_worst
b1_5 = calc_b1(data['radius_worst'], data['area_worst'])
b0_5 = calc_b0(data['radius_worst'], data['area_worst'])

print(f"y = {b0_5:0.3f} + {b1_5:0.3f}x")
```

```python
def calc_regression(X, Y, Ax):
    b1 = calc_b1(X,Y)
    b0 = calc_b0(X,Y)

    xplot = np.linspace(X.min(), X.max(), 100)

    yplot = b0 + b1 * xplot

    sns.scatterplot(x=X, y=Y, ax=Ax)
    sns.scatterplot(x=xplot, y=yplot, ax=Ax, s=10)
```

```python
fig, ax = plt.subplots(5, 1, figsize=(10, 50))

calc_regression(data['radius_mean'], data['perimeter_mean'], ax[0])
calc_regression(data['perimeter_worst'], data['radius_worst'], ax[1])
calc_regression(data['area_mean'], data['radius_mean'], ax[2])
calc_regression(data['area_mean'], data['perimeter_mean'], ax[3])
calc_regression(data['radius_worst'], data['area_worst'], ax[4])


ax[0].set_title('Average radius vs the average perimeter')
ax[1].set_title('Worst perimiter vs the worst radius')
ax[2].set_title('Average area vs the average radius')
ax[3].set_title('Average area vs the average perimeter')
ax[4].set_title('Worst radius vs the worst area')
```
