```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(0, 1, 101)
y = x**2 + np.random.randn(len(x))/50
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left bottom width height
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left bottom width height

ax.plot(x,y, c='lightblue', label='Original')
ax.plot(x, x**2, c='yellow', linestyle='dotted', label='$y=x^2$')

plt.title('Random data follwing the $y=x^2$ line')
plt.xlabel('X Value')
plt.ylabel('Y Value')

plt.legend(loc='lower right')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left bottom width height

ax.plot(x,y, c='lightblue', label='Original')
ax.plot(x, x**2, c='yellow', linestyle='dotted', label='$y=x^2$')

plt.title('Random data follwing the $y=x^2$ line')
plt.xlabel('X Value')
plt.ylabel('Y Value')

ax_inset = fig.add_axes([0.2, 0.6, 0.3, 0.25])
ax_inset.plot(x,y, c='lightblue', label='Original')
ax_inset.plot(x, x**2, c='yellow', linestyle='dotted', label='$y=x^2$')
ax_inset.set_xlim([0.5, 0.6])
ax_inset.set_ylim([0.25, 0.4])

ax_inset.set_xticks([0.5, 0.52, 0.54, 0.56, 0.58, 0.6])
ax_inset.set_xticklabels(['Min', '0.52', '0.54', '0.56', '0.58', 'Max'])

ax.legend(loc='lower right')
plt.show()
```

# Seaborn

```python
import seaborn as sns 

tips = sns.load_dataset('tips')
tips
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.scatter(tips[tips['time'] == 'Lunch']['total_bill'], 
           tips[tips['time'] == 'Lunch']['tip'],
           c='Orange')

plt.title('Tips as a function of the Total Bill')
plt.xlabel('Total Bill ($US)')
plt.ylabel('Tip Amount ($US)')

plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', size='size', alpha=0.7)

plt.title('Tips as a function of the Total Bill')
plt.xlabel('Total Bill ($US)')
plt.ylabel('Tip Amount ($US)')

plt.show()
```
