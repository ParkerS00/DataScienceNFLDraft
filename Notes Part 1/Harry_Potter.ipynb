```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/drolsonmi/math3080/main/Datasets/HarryPotterCharacters.csv", delimiter=';')
data
```

```python
data.describe()
data.value_counts()
data.info()
```

```python
fig, ax = plt.subplots(1,3, figsize=(15, 4))

sns.countplot(data=data, x='Gender', ax=ax[2])
ax[2].set_title('Gender')

sns.boxplot(data=data, x='Birth', y='Death', ax=ax[0])
ax[0].set_title('Birth vs Death')

sns.histplot(data=data, x='Eye colour', ax=ax[1])
ax[1].set_title('Eye Color')
```

```python
fig, ax = plt.subplots(1,3, figsize=(15, 4))

sns.scatterplot(data=data, x='House', y='Eye colour', ax=ax[0], hue='Loyalty')
ax[0].set_xticklabels(['Gryf', 'Rav', 'Sly', 'Huff', 'Magic', 'Drum'])
```
