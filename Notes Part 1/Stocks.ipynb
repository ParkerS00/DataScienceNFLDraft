```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

data = pd.read_csv("https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=IBM&outputsize=compact&apikey=ENIPDF3XPHW9IUUE&datatype=csv", delimiter=',')
data
```

```python
data.info()
data.describe()
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.scatter(data['open'], data['close'], c='Orange')

plt.title('Low of the Stock vs the High of the Stock')
plt.xlabel('Low of stock')
plt.ylabel('High of stock')

plt.show()
```
