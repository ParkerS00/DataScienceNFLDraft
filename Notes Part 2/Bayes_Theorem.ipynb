# Bayes Theorem
- P(A) = (# of desired outcomes) / (total # of outcomes)
- P(B | A) = (# of desired outcomes) / (total outcomes in A)
- P(A and B) = P(A) * P(B | A)
- P(B | A) = ( P(B) * P(A | B) ) / P(A) *The Actual Bayes Theorem*
- Belief is not determined by one event, instead each piece of information updated the belief.
- The accumulation of information led to a belief.
- P(Belief | Info) = ( P(Belief) * P(Info | Belief) ) / P(Info)
- P(Hypothesis | Evidence) = ( P(Hypothesis) * P(Evidence | Hypothesis) ) / P(Evidence)
