# Pivot Tables & Melting
- Long format: basic and few columns such as an id and the var1, var2, var3 with many rows spanning down the columns.
- Wide format: take the long format and flip it on its side so that your rows are var1, you columns are var2, and the values inside the table are var3, common at times to lose the id when you are switching from long to wide because you don't need to know which id you are looking at any more. 
- Melting: turns the wide format back into a long format, in this one you take the 3 variables and put them back into a vertical table with many rows and give you things like Index, Column, Value where index would be your var1, column would be your var2, and the value would be var3 still not gaining back the id.
- Index and columns are usually categorical variables

```python
import pandas as pd
data = pd.DataFrame({
    'Student' : ['Kyler', 'Zack', 'Aaron', 'Ethan', 'Parker', 'Seth', 'Dusty'],
    'Exam1' : [84, 74, 99, 85, 97, 60, 91],
    'HW1' : ['A', 'B', 'B', 'A', 'C', 'A', 'A'],
    'Grade' : ['A', 'B', 'A', 'B', 'B', 'C', 'A']
})

data
```

```python
data_table = pd.pivot_table(data, values='Exam1', index='HW1', columns='Grade')
data_table
```

```python
pd.melt(data_table, id_vars='A')
```

# One hot-encoding, ordinal encoding
- Turning a categorical variable into a numeric variable so that you can work with it

```python
order = {
    'A' : 1,
    'B' : 2,
    'C' : 3
}

data['NumGrade'] = data['Grade'].map(order)
data
```

```python
pd.get_dummies(data['Grade'])
```

# Cleaning Data
- Interpolate: Take the value before and after and then take either the min,max, mean, median or others.
- Forward-Fill (ffill): take the value before and then put it to the one in front.
- Back-Fill (bfill): take the value in front and put it for the value behind it. 
- Fill: give it some sort of a default value, often this is the average of the whole column.
- Drop empty data

# Outliers
- If you have a large amount of data you don't need to worry about it because they aren't going to affect the averages or anything all that much
- If you have a small amount of data you can drop the outliers so they won't affect the averages 
- Q1 - 1.5IQR, Q2 + 1.5IQR
